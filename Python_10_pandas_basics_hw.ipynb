**Задание 1**

Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.


```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
```

```python
star_films = ratings[ ratings.rating == 5.0 ].groupby('movieId').count().sort_values('rating', ascending=False)
star_films_filtered = star_films.filter(items = ['movieId', 'rating'])
star_films_filtered = star_films_filtered.rename(columns={'rating': 'Number_5*'})
joined = star_films_filtered.merge(movies, on='movieId', how='left')
print('Три фильма с наибольшим количеством оценок 5.0')
joined.filter(items = ['movieId', 'title', 'Number_5*']).head(3)

#Достаточно много строк получилось. Можно ли такой же результат получить более коротким путем?
```

**Задание 2**

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.



```python
power = pd.read_csv('power.csv')
```

```python
baltic_countries = power[ (power['country']=='Lithuania') | (power['country']=='Latvia') | (power['country']=='Estonia') ]
category_filter = baltic_countries[ (baltic_countries['category']==4) | (baltic_countries['category']==12) | (baltic_countries['category']==21)]
year_filter = category_filter[ (category_filter['year'] >= 2005) & (category_filter['year'] <= 2010)]
quantity_filter = year_filter[ (year_filter['quantity']) >= 0 ]
sum_quantity = quantity_filter['quantity'].sum()

print('Суммарное потребление электричества -', sum_quantity)
print('Проверка таблицы данных:')
quantity_filter.sort_values('quantity', ascending=True)

#Получился длинный код. Можно ли как-то его оптимизировать?
```
